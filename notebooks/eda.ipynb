```python
# Core data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Date/time handling
from datetime import datetime

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
df = pd.read_csv("data/202409_citibike_tripdata_combined.csv")

df.head()
```

```python
df["started_at"] = pd.to_datetime(df["started_at"])
df["ended_at"] = pd.to_datetime(df["ended_at"])

df["ride_duration_min"] = (
    df["ended_at"] - df["started_at"]
).dt.total_seconds() / 60

df.head()
```

```python
df = df.dropna()

df.isna().sum()
```

```python
# Round UP to nearest minute
df["ride_minutes_charged"] = np.ceil(df["ride_duration_min"]).astype(int)

def compute_price(row):
    minutes = row["ride_minutes_charged"]
    member_type = row["member_casual"]   # "casual" or "member"
    rideable = str(row["rideable_type"]).lower()

    # Included ride time
    included = 30 if member_type == "casual" else 45

    # Base per-minute rate
    if "electric" in rideable:
        base_rate = 0.38 if member_type == "casual" else 0.25
    else:
        base_rate = 0.0  # classic/docked bikes

    within = min(minutes, included)
    over = max(minutes - included, 0)

    # Pricing logic
    price = base_rate * within + (base_rate + 0.25) * over

    return round(price, 2)

# Apply pricing
df["ride_price"] = df.apply(compute_price, axis=1)


df.head()
```

```python
os.makedirs("data", exist_ok=True)

# Save cleaned dataset
df.to_csv("data/202409_citibike_tripdata_cleaned.csv", index=False)
```

```python
# Create subfolder
output_dir = "data/samples"
os.makedirs(output_dir, exist_ok=True)

# Randomly sample 20,000 rows (reproducible)
df_sample = df.sample(n=20000, random_state=42)

# Save sample
df_sample.to_csv(
    os.path.join(output_dir, "202409_citibike_tripdata_cleaned_sample_20000_random.csv"),
    index=False
)
```
